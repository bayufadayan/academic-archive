```python
import requests #Modul request ini memperbolehkan kita untuk meminta HTTP Request melalui Python
import pandas as pd # Python Data Analysis library spt biasa
import seaborn as sns # ini sama kayak matplotlib, untuk visualisasi statistik data
import matplotlib.pyplot as plt # ^
from bs4 import BeautifulSoup as bs4 # Library untuk pengambil(PARSING) data dari HTML dan XML
```

```python
sekolah_req = requests.get("https://top-1000-sekolah.ltmpt.ac.id").text

sekolah_req
```

```python
sekolah_soup = bs4(sekolah_req,"lxml")
```

```python
sekolah_table = sekolah_soup.find('table', {'class' : 'table table-striped table-bordered'})
sekolah_table
```

```python
sekolah_df = pd.read_html(str(sekolah_table))
sekolah_df
```

```python
sekolah_df = sekolah_df[0]
sekolah_df
```

Membagusi nama kolom nya

```python
# mengganti nama kolom agar hanya memiliki satu baris
sekolah_df = sekolah_df.rename(columns={
    'Unnamed: 0_level_0': 'Maaf Ka',
    'Unnamed: 1_level_0': '.',
    'Unnamed: 2_level_0': 'Saya Telat Banget',
    'Unnamed: 3_level_0': 'Semoga nilainya bagus',
    'Unnamed: 4_level_0': 'Semoga rezekinya lancar',
    'Unnamed: 5_level_0': 'Aamiin',
    'Unnamed: 6_level_0': 'Cepat dapat momongan'
})

# menghapus baris header kedua
sekolah_df.columns = sekolah_df.columns.droplevel(1)
```

```python
sekolah_df
```

Menghapus kolom "." karena tidak diperlukan

```python
sekolah_df = sekolah_df.drop(columns=['.'])
```

```python
sekolah_df
```

Mengganti nama salah satu field, saya akan mengganti field Sekolah

```python
sekolah_df = sekolah_df.rename(columns={"Sekolah" : "065121100"})
sekolah_df
```
