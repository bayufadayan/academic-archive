```python
import pandas as pd
```

```python
data = pd.read_csv('Pengaruh Kualitas Tidur dengan Tingkat Produktivitas Sehari-hari (Responses) - Form Responses 1.csv')
data
```

## Hapus Kolom TimeStamp

```python
df = data.drop(columns=['Timestamp'])
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'Usia anda saat ini\n*contoh : 13 Tahun' : 'umur'})
df = df.rename(columns={'Jenis Kelamin' : 'gender'})
df = df.rename(columns={'Status Anda saat ini (Anda dapat memilih lebih dari satu opsi)' : 'status'})
df = df.rename(columns={'Durasi anda tidur dalam sehari' : 'durasi tidur'})
df = df.rename(columns={'Pilihlah sesuai dengan kebiasaan Anda [Apakah Anda rutin mengonsumsi kafein?]' : 'rutinitas konsumsi kafein'})
df = df.rename(columns={'Pilihlah sesuai dengan kebiasaan Anda [Apakah Anda rutin berolahraga?]' : 'rutinitas berolahraga'})
df = df.rename(columns={'Pilihlah sesuai dengan kebiasaan Anda [Apakah Anda sering mengonsumsi makanan atau minuman sebelum tidur?]' : 'rutinitas konsumsi makanan dan minum sebelum tidur'})
df = df.rename(columns={'Pilihlah sesuai dengan kebiasaan Anda [Apakah Anda memiliki kebiasaan menggunakan perangkat elektronik (seperti smartphone atau tablet) sebelum tidur?]' : 'rutinitas menggunakan piranti elektronik'})
df = df.rename(columns={'Pilihlah sesuai dengan kebiasaan Anda [Apakah Anda memiliki rutinitas tidur yang teratur, seperti menjaga jadwal tidur yang konsisten?]' : 'rutinitas tidur dengan teratur'})
df = df.rename(columns={'Pilihlah sesuai dengan produktivitas Anda [Seberapa sering Anda merasa segar dan bertenaga saat bangun tidur?]' : 'merasa segar saat bangun tidur'})
df = df.rename(columns={'Pilihlah sesuai dengan produktivitas Anda [Seberapa sering Anda mengalami kantuk berlebihan atau mengantuk di siang hari?]' : 'merasa kantuk disiang hari'})
df = df.rename(columns={'Pilihlah sesuai dengan produktivitas Anda [Seberapa sering Anda mencapai target dalam pekerjaan Anda?]' : 'mencapaian target dalam pekerjaan'})
df = df.rename(columns={'Pilih sesuai dengan diri Anda sendiri [Apakah Anda pernah merasa sulit untuk menyelesaikan tugas atau pekerjaan karena kurang tidur?]' : 'merasa sulit menyelesaikan tugas karena kurang tidur'})
df = df.rename(columns={'Pilih sesuai dengan diri Anda sendiri [Apakah Anda merasa memiliki cukup energi untuk menyelesaikan tugas atau pekerjaan yang ada?]' : 'merasa memiliki energi yang cukup untuk menyelesaikan pekerjaan'})
df = df.rename(columns={'Pilih sesuai dengan diri Anda sendiri [Apakah Anda merasa terorganisir dalam mengelola waktu dan tugas-tugas Anda?]' : 'merasa terorganisir dalam mengelola waktu'})
df = df.rename(columns={'Pilih sesuai dengan diri Anda sendiri [Apakah Anda merasa lebih produktif saat tidur dengan baik dan cukup dibandingkan saat tidur yang kurang berkualitas?]' : 'merasa produktif saat cukup tidur'})
df = df.rename(columns={'Apa penyebab yang membuat anda mengalami kesulitan untuk tidur?' : 'penyebab kesulitan untuk tidur'})
df = df.rename(columns={'Apa faktor utama yang dapat membuat Anda menjadi kurang produktif/malas?' : 'faktor utama kurang produktif'})
df.head()
```

## Membersihkan Data

### Kolom umur

```python
df_umur = df['umur']
df_umur.head()
```

```python
df_umur = df_umur.str.lower()
df_umur.head()
```

```python
df_umur = df_umur.str.replace('tahun', '')\
                  .str.replace('thn', '')\
                  .str.replace(' ','')
df_umur
```

### Status

```python
df_status = df['status'].str.lower()
df_status.head()
```

```python
df_status = df_status.str.replace('pelajar \(sd, smp, sma\)', 'pelajar', regex=True)
df_status = df_status.str.replace('pekerja full-time \(karyawan tetap, karyawan kontrak, dll.\)', 'pekerja full-time', regex=True)
df_status = df_status.str.replace('pengusaha \(wirausaha atau pemilik usaha\)', 'pengusaha', regex=True)
df_status = df_status.str.replace('tidak bekerja \(pengangguran, mencari pekerjaan, pensiun dini, dll.\)', 'tidak bekerja', regex=True)

df_status
```

```python
df_status = df_status.str.split(', ')
```

```python
df_status
```

```python
df_status_count = df_status.apply(pd.Series).stack().value_counts()
df_status_count
```

```python
df_durasitidur = df['durasi tidur'].value_counts()
df_durasitidur
```

```python
df_penyebab = df['penyebab kesulitan untuk tidur']
df_penyebab
# ini mah nyoba
```

```python
df_clean = pd.concat([df_umur, df['gender'], df['status'], df['durasi tidur'], df['rutinitas konsumsi kafein'], df['rutinitas berolahraga'], df['rutinitas konsumsi makanan dan minum sebelum tidur'], df['rutinitas menggunakan piranti elektronik'], df['rutinitas tidur dengan teratur'], df['merasa segar saat bangun tidur'], df['merasa kantuk disiang hari'], df['mencapaian target dalam pekerjaan'], df['merasa sulit menyelesaikan tugas karena kurang tidur'], df['merasa memiliki energi yang cukup untuk menyelesaikan pekerjaan'], df['merasa terorganisir dalam mengelola waktu'], df['merasa produktif saat cukup tidur'], df['penyebab kesulitan untuk tidur'], df['faktor utama kurang produktif']], axis=1)
df_clean
```

```python
kolom = ['umur', 'gender', 'status', 'durasi tidur', 'rutinitas konsumsi kafein',
       'rutinitas berolahraga',
       'rutinitas konsumsi makanan dan minum sebelum tidur',
       'rutinitas menggunakan piranti elektronik',
       'rutinitas tidur dengan teratur', 'merasa segar saat bangun tidur',
       'merasa kantuk disiang hari', 'mencapaian target dalam pekerjaan',
       'merasa sulit menyelesaikan tugas karena kurang tidur',
       'merasa memiliki energi yang cukup untuk menyelesaikan pekerjaan',
       'merasa terorganisir dalam mengelola waktu',
       'merasa produktif saat cukup tidur', 'penyebab kesulitan untuk tidur',
       'faktor utama kurang produktif']

for i in range(0, len(kolom)):
  print(f"df['{kolom[i]}'],", end=" ")
```
