Pengambilan data

```python
data = []

for page in range(167, 220):
    url = f'https://www.detik.com/search/searchall?query=kecelakaan+lalu+lintas&siteid=2&sortby=time&sorttime=3&fromdatex=01/04/2022&todatex=31/03/2023&page={page}'
    r = req.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    for article in soup.find_all('article'):
        tautan = article.find('a')['href']
        caption = article.find('p').text
        response = req.get(tautan)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Mengambil judul artikel
        judul = soup.find('h1', {'class': 'detail__title'})
        if judul:
          judul = judul.text
          judul = judul.replace('\n','')
        else :
          judul = None

        # Mengambil tanggal publikasi
        tanggal = soup.find('div', {'class': 'detail__date'})
        if tanggal:
          tanggal = tanggal.text
        else :
          tanggal = tanggal

        # Mengambil Penulis Artikel
        writer = soup.find('div', {'class': 'detail__author'})
        if writer:
            writer = writer.text
        else:
            writer = None

        # Mengambil label
        label = soup.find('span', {'class': 'detail__label'})
        if label:
            label = label.text
        else:
            label = None

        # Mengambil kota
        city = soup.find('strong')
        if city:
            city = city.text
        else:
            city = None

        # Mengambil descripsi lengkap
        content = soup.find('div', {'class': 'detail__body-text'})
        if content :
          content = content.text
          content = content.replace('\n', '').replace('\r','').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT', '')
        else :
          content = None

        # Menambahkan data ke list
        data.append([judul, tanggal, writer, label, city, content, tautan])

columns = ['Judul', 'Tanggal', 'Penulis', 'Label', 'Kota', 'Deskripsi', 'Url']
df = pd.DataFrame(data, columns=columns)
print(df)
```

## **Kompas.Com Scaraping ()**

```python
data_kompas = []

for page in range (3, 28) :
  url = f'https://www.kompas.com/tag/kecelakaan-lalu-lintas?page={page}'
  r = req.get(url)
  soup = BeautifulSoup(r.content, 'html.parser')
  articles = soup.find_all("div", class_="article__list clearfix")
  for article in articles:
    tautan = article.find('a')['href']
    response = req.get(tautan)
    soup = BeautifulSoup(response.content, 'html.parser')

    judul = soup.find("h1", class_="read__title")
    if judul:
        judul = judul.text
        tanggal = soup.find("div", class_="read__time").text
        writer = soup.find("div", class_="read__credit__item").find("a").text
        label = soup.find("ul", class_="breadcrumb__wrap")
        kategori = label.find_all("a")[-1].text.strip()
        city = soup.find("div", class_="read__content").find("strong").text
        if city:
            city.strip()
        else :
            city2 = soup.find("div", class_="read__content").find("a").text
            city = city2
        content = soup.find("div", class_="read__content").text
    else :
        judul = None
        tanggal = None
        writer = None
        kategori = None
        city = None
        content = None

    # Menambahkan data ke list
    data_kompas.append({'Judul': judul, 'Tanggal': tanggal, 'Penulis': writer, 'Label': kategori, 'Kota': city, 'Content': content, 'Url': tautan})

    print(tanggal)
```

### **Basic Data Cleaning (deskripsi)**

```python
df_desc = df3['deskripsi']
df_desc
```

```python
panjang_kata = df_desc.apply(len)
panjang_kata
```

```python
import re # re ini regular expression yang tadi di atas di jelasin
def remove_html(text): #define variable fungsi remove html/html tag dalam bentuk dari variable apapun yang berbentuk text
    text = text.replace("\n"," ") #yang ini tiap ada baris baru pake /n di ganti pake spasi aja yang ada di var text
    pattern = re.compile('<.*?>') #semua HTML tags
    return pattern.sub(r'', text) #return
```

```python
df_desc_processed = df_desc.apply(remove_html)
print(df_desc_processed)
```

```python
def remove_headings(text):
    pattern = re.compile('\w+:')
    return pattern.sub(r'', text)
```

```python
df_desc_processed = df_desc.apply(remove_headings)
print (df_desc_processed)
# liat EXAM : <~ ini adalah heading yang hilang dari contoh sebelumnya
```

```python
def replace_mult_spaces(text):
    text = text.replace("&quot","")
    pattern = re.compile(' +')
    text = pattern.sub(r' ', text)
    text = text.strip()
    return text

def replace_other_chars(text):
    pattern = re.compile(r'[()!@&;#]')
    text = pattern.sub(r'', text)
    return text
```

```python
df_desc_processed = df_desc.apply(replace_mult_spaces)
df_desc_processed = df_desc.apply(replace_other_chars)
print (df_desc_processed)
```

```python
def clean_text(text):
    text = remove_html(text)
    text = remove_headings(text)
    text = replace_mult_spaces(text)
    text = replace_other_chars(text)
    text = text.lower()
    return text
```

```python
df_desc_processed_clean = df_desc.apply(clean_text)
print(df_desc_processed_clean)
```

```python
panjang_kata_new = df_desc_processed_clean.apply(len)
print("Sebelum : ")
print(panjang_kata)
print("Sesudah : ")
print(panjang_kata_new)
```

```python
import pandas as pd
import re

vehicle_keywords = ['mobil', 'motor', 'truk', 'pesawat', 'kapal', 'bus', 'kereta', 'perahu']

# 3. Definisikan fungsi untuk mencari kendaraan pada deskripsi
def find_vehicles(description):
    vehicles = []
    for keyword in vehicle_keywords:
        if re.search(keyword, description, re.IGNORECASE):
            vehicles.append(keyword)
    return vehicles
```
