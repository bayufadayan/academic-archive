# Membuat kamus yang merepresentasikan hubungan antara gejala dan penyakit

```python
gejala_sakit = {
    'Demam': ['Flu', 'Malaria'], #jika gejala demam ada kemungkinan dua penyakit flu atau malaria
    'Batuk': ['Flu'], #jika gejala batuk ada kemungkinan terkena flu
    'Mual': ['Malaria'], #jika gejala mual ada kemungkinan terkena malaria
    'Nyeri Perut': ['Malaria'] #jika gejala mual ada kemungkinan terkena malaria
}
```

```python
def diagnosa(gejala_sakit, gejala):
    # Inisialisasi himpunan penyakit terkait yang mungkin diderita pasien
    penyakit_terkait = set()

    # Iterasi melalui setiap gejala yang dimasukkan pengguna
    for gejala_saat_ini in gejala:
        # Periksa apakah gejala tersebut terkait dengan penyakit dalam semantic network
        if gejala_saat_ini in gejala_sakit:
            # Jika terkait, tambahkan semua penyakit terkait ke dalam himpunan penyakit_terkait
            penyakit_terkait.update(gejala_sakit[gejala_saat_ini])

    # Jika terdapat penyakit_terkait yang ditemukan, kembalikan himpunan tersebut
    if penyakit_terkait:
        return penyakit_terkait
    # Jika tidak ditemukan penyakit terkait, kembalikan pesan kesalahan
    else:
        return "Maaf, tidak ada informasi penyakit yang sesuai dengan gejala yang Anda sebutkan."
```

```python
# Contoh penggunaan sistem pakar
gejala = ['Demam', 'Batuk']
print("Hasil Diagnosa:", diagnosa(gejala_sakit, gejala))
```
