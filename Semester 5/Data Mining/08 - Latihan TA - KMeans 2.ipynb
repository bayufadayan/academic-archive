```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dfmhs = pd.read_csv('nilaimahasiswa.csv', sep = ';', engine='python')
```

```python
dfmhs.describe()
```

```python
dfmhs.info()
```

```python
plt.scatter(dfmhs['uts'], dfmhs['uas'])
plt.xlabel("Uts")
plt.ylabel("Uas")
plt.title('Grafik Nilai')
plt.show()
```

```python
X_train = dfmhs[['uts', 'uas']].values
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters = 2)
kmean
```

```python
y_cluster = kmean.fit_predict(X_train)
```

```python
y_cluster
```

```python
dfmhs['Cluster'] = y_cluster
```

```python
dfmhs
```

```python
kmean.cluster_centers_
```

```python
plt.figure(figsize = (10,5))
plt.scatter(X_train[:,0], X_train[:, 1], c = kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0],
            kmean.cluster_centers_[0,1],
            cmap= 'rainbow',
            marker = '*',
            s = 150,
            label = 'Centroid 0')

plt.scatter(kmean.cluster_centers_[1,0],
            kmean.cluster_centers_[1,1],
            cmap= 'rainbow',
            marker = '*',
            s = 150,
            label = 'Centroid 1')

plt.xlabel('UTS')
plt.ylabel('UAS')
plt.legend()
```

```python
# Elbow method
inertias = []
k_range = range(1, 10)
for k in k_range:
  km = KMeans(n_clusters = k).fit(X_train)
  inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.xlabel('k')
plt.ylabel('Sum of Error')
plt.plot(k_range, inertias)
plt.grid()
```
