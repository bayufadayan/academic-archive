#### Indexing dan Selecting

```python
# Untuk row tertentu berdasarkan indexnya saja
df.iloc[1]
```

```python
# 0:2 berarti akan mengambil baris dengan index ke 0 sampai 1 (2 baris)
# 3:9 berarti akan mengambil baris dengan index ke 3 sampai 8 (6 baris)
# 3 diakhir adalah kolom
df.iloc[3:9, 2]
```

```python
# Mengambil index ke 3 sampai 8 dengan semua kolom
df.iloc[3:9, :]
```

```python
# untuk kolom sama saja seperti baris
df.iloc[3:9, 3:7]
```

```python
# Mau cari index ke 33 di kolom Luas RTH (km2)(B)
df.loc[33, 'Luas RTH (km2)(B)']
```

```python
# Ngambil sampai index ke 6, untuk kolom tertentu
df.loc[:6, ['Luas RTH (km2)(B)', 'Taman Kota (km2)']]
```

```python
# Untuk meng-set menggunakan nama dari isi kolom
df.set_index('Provinsi')
```

```python
# Slicing DataFrame secara Conditional
df['Tahun'] == 2022
```

```python
df[
    (df['Taman Kota (km2)'] > 3) & (df['Tahun'] == 2022)
]
```

```python
filter = df['% RTH(B/A)'] > 0.1
df.where(filter)
```

#### **Merging**

```python
dataframe_1 = pd.DataFrame(
    [
        ['a', 1],
        ['b', 2]
    ],
    columns = ['huruf', 'angka']
)

dataframe_1
```

```python
dataframe_2 = pd.DataFrame(
    [
        ['c', 3],
        ['d', 4]
    ],
    columns = ['huruf', 'angka']
)

dataframe_2
```

```python
# axis = 0 = nambah kebawah / nambah breakaris
# axis = 1 = nambah kesampig / nambah kolom
pd.concat([dataframe_1, dataframe_2], axis = 0)
```

## Numpy
berhubungan dengan array list
memberi dimensi ke array

```python
mylist = [1,2,3]
np.array(mylist)
```

```python
mymatrix = [[1,2,3], [4,5,6]]
np.array(mymatrix)
```

```python
# ingin membuat array 1 dimensi dalam nilai terurut dengan interval tertentu
# (0, 11, 2) :
# 0 : Angka mulainya
# 11 : Masimal index
# 2 : interval
newlist = np.arange(0, 11, 2)
newlist
```

```python
# ingin membuat array 1 dimensi dalam nilai terurut dengan interval tertentu
# (0, 11) :
# 0 : Angka mulainya
# 11 : Masimal index
newlist_2 = np.arange(0, 11)
newlist_2
```

```python
# berdasarkan pembagian
# dari 0 sampai 100 dibagi menjadi 5 hasilnya.
np.linspace(0, 100, 5)
```
