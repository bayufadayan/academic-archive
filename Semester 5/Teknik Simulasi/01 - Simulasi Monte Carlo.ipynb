```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulasi Pelemparan Dadu

```python
def monte_carlo_simulation(num_simulation):
  results = []
  for _ in range(num_simulation):
    result = np.random.randint(1,7)
    results.append(result)
  return results
```

```python
num_simulation = 1000
simulation_results = monte_carlo_simulation(num_simulation)
```

```python
plt.hist(simulation_results, bins=np.arange(0.5, 7.5,1), edgecolor='black', align='mid')
plt.title('Simulasi Pelemparan Dadu')
plt.xlabel('Nilai Dadu')
plt.ylabel('Frekuensi')
plt.show()
```

# Simulasi Pengiriman Paket

```python
def delivery_simulation(num_simulation, mean_traffic, stdev_traffic, mean_weather, stdev_weather):
  delivery_times = []
  for _ in range(num_simulation):
    traffic_delay = np.random.normal(mean_traffic, stdev_traffic)
    weather_delay = np.random.normal(mean_weather, stdev_weather)
    delivery_time = 60 + traffic_delay + weather_delay
    delivery_times.append(delivery_time)
  return delivery_times
```

```python
num_simulation = 1000
mean_traffic = 5
stdev_traffic = 2
mean_weather = 3
stdev_weather = 1
```

```python
delivery_times = delivery_simulation(num_simulation, mean_traffic, stdev_traffic, mean_weather, stdev_weather)
```

```python
plt.hist(delivery_times, bins=30, edgecolor='black')
plt.title('Simulasi Pengiriman Paket')
plt.xlabel('Waktu Pengiriman Paket')
plt.ylabel('Frekuensi')
plt.show()
```
