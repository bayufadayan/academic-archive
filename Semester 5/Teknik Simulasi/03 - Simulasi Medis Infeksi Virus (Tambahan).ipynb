```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = False
```

Deklarasi Variable

```python
kr1 = 1e5 #var h
kr2 = 0.1 #var i
kr3 = 2e-7 #var v // liat ppt
kr4 = 0.5 # banyaknya var h yang di konversi menjadi i, berarti 50%
kr5 = 5 # banyaknya var v yang mati pertahun
kr6 = 100 # banyaknya var i yang menjadi v
```

Define function perhitungan simulasi

```python
def HIV(x,t): #assign variable dan perhitungan berdasarkan waktu x = cells, t = waktu
  h = x[0]
  i = x[1]
  v = x[2]
  dhdt = kr1 - kr2*h - kr3 * h * v #perhitungan cell sehat
  didt = kr3 * h * v - kr4 * i #perhitungan cell yang terinfeksi
  dvdt = -kr3 * h * v - kr5 * v + kr6 * i #perhitungan cell terinfeksi yang menjadi virus
  return[dhdt, didt, dvdt]
```

define nilai awal dan interval waktu

```python
x0 = [1e6, 0, 100] # h , i dan v
t = np.linspace(0,15,100)
```

```python
x = odeint(HIV, x0, t) #menjalankan variable x0 kedalam fungsi HIV ditambah interval t
```

```python
plt.semilogy(t, x[:,0], label=r'$H$')
plt.semilogy(t, x[:,1], label=r'$I$')
plt.semilogy(t, x[:,2], label=r'$V$')
plt.xlabel(r'Waktu (/tahun)')
plt.ylabel(r'Cell')
plt.legend(frameon=False)
plt.show()
```

Di grafik ini bisa dilihat, dari simulasi yang dilakukan selama 14 tahun, cell sehat perlahan akan di infeksi oleh virus, dan cell yang terinfeksi akan membuat populasi cell virus bertambah tiap tahun nya, sehingga membuat cell sehat berkurang.<br><br> Ini adalah contoh penggunaan teknik simulasi dalam bidang medis, perhitungan tentu dibutuhkan rumus yang sudah paten untuk prediksi, dan variable harus menggunakan data yang telah teruji
