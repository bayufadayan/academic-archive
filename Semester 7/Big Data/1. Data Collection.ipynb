**Melakukan Pencocokan Kolom antara Komoditas Publikasi dan Komoditas Umum dengan Fuzzy Matching**

```python
!pip install fuzzywuzzy
!pip install python-Levenshtein
```

```python
from fuzzywuzzy import fuzz
```

```python
def match_commodities(komoditas_umum, komoditas_publikasi):
    matched = []
    for k in komoditas_umum:
        for p in komoditas_publikasi:
            if fuzz.partial_ratio(k.lower(), p.lower()) > 90:  # Match jika similarity > 90%
                matched.append((k, p))
    return matched
```

```python
matched_commodities = match_commodities(komoditas_umum, komoditas_publikasi)
matched_commodities
```

Terlihat bahwa hanay 5 kolom yang terindikasi mirip dengan kolom publikasi umum, yang berarti komoditas tersebut tergolong penting dan lebih baik tidak dihilangkan.

Penting berarti melambangkan prioritas dalam kolom. Oleh karena itu disini kami akan memilih kolom sesuai dengan publikasi umum dengan kolom prioritas dari matched_comodities
